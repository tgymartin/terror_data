```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns
sns.set(style='dark')
```

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv', delimiter=',', low_memory = False)
```

```python
# 20 MOST FREQUENT TARGETS (SINCE 1970)

all_targets = (df[df['iyear'].dropna() >= 2012]
           .groupby(['targsubtype1_txt'])
           .size()
           .reset_index()
           .sort_values([0], ascending=False)
           .head(20))

ax = sns.barplot(x='targsubtype1_txt', y=0, data=all_targets)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set(xlabel='Type', ylabel='Occurrences')

plt.show()
```

```python
# CHANGE IN FREQUENCY (PER YEAR) OF ATTACKS ON URBAN TARGETS OVER TIME

urban_targets = ['School/University/Educational Building', 'Construction', 'Place of Worship', 'Hotel/Resort', 'Government Building/Facility/Office', 'House/Apartment/Residence']
short_labels = ['School', 'Construction Site', 'Place of Worship', 'Hotel', 'Government Building', 'Residence']

for i in range(len(urban_targets)):
    table = df[(df['targsubtype1_txt'] == urban_targets[i])]
    table = (table[['iyear', 'targsubtype1_txt']]
             .groupby(['iyear', 'targsubtype1_txt'])
             .size()
             .reset_index())
    
    ax = sns.lineplot(x='iyear', y=0, data=table, label=short_labels[i])
    
ax.set(xlabel='Year', ylabel='Occurrences')

plt.show()
```
